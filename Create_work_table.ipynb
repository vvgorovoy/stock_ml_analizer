Создаем, собственно, рабочую таблицу

```python
def add_5_names(st):
    for i in np.arange(1,6):
        cols.append(st + ' ' + str(i))

cols = splist.columns.tolist() + ['Year of examination']
add_5_names('Revenue')
add_5_names('Net Income')
add_5_names('EBITDA')
add_5_names('Cash')
add_5_names('Debt')
add_5_names('Total Assets')
add_5_names('Total Liabilities')
add_5_names('Market Cap')
add_5_names('Price')
cols
```

Переносим все первичные финансовые данные, а также капитализации и цены в рабочую таблицу

```python
def get_row_of_five(i, c):
    row5 = [fin_data.iloc[i, c], fin_data.iloc[i, c+1], fin_data.iloc[i, c+2], fin_data.iloc[i, c+3], fin_data.iloc[i, c+4]]
    return row5

def add_fins_to_row(row, j, i):
    c_rev = 16 - j
    revs = get_row_of_five(i, c_rev)
    c_inc = 32 - j
    incs = get_row_of_five(i, c_inc)
    c_ebitda = 48 - j
    ebitdas = get_row_of_five(i, c_ebitda)
    c_cash = 64 - j
    cashes = get_row_of_five(i, c_cash)
    c_debt = 80 - j
    debts = get_row_of_five(i, c_debt)
    c_ass = 96 - j
    assets = get_row_of_five(i, c_ass)
    c_liab = 112 - j
    liabs = get_row_of_five(i, c_liab)
    c_mcaps = 128 - j
    mcaps = get_row_of_five(i, c_mcaps)
    c_prices = 144 - j
    prices = get_row_of_five(i, c_prices)
    new_elements = revs + incs + ebitdas + cashes + debts + assets + liabs + mcaps + prices
    row = row + new_elements
    return row
```

```python
n = 0
for i in np.arange(len(fin_data.index)):
    for j in np.arange(2017-first_years[i],5,-1):
        row = [splist.Company[i], splist.Ticker[i], splist.Sector[i], splist['Sub-sector'][i], f'{int(2022-j)}']
        row = add_fins_to_row(row, int(j), i)
        work_table.loc[n] = row
        n+=1
for i in np.arange(len(fin_data.index)):
    row = [splist.Company[i], splist.Ticker[i], splist.Sector[i], splist['Sub-sector'][i], f'{int(2021)}']
    row = add_fins_to_row(row, 1, i)
    work_table.loc[n] = row
    n+=1    
```

Расчитываем значения мультипликаторов для всех компаний

```python
c_rev = 5
c_inc = 10
c_ebitda = 15
c_cash = 20
c_debt = 25
c_tass = 30
c_tliab = 35
c_mcap = 40
c_price = 45
c_pe = 50
c_ps = 55
c_pb = 60
c_ev_ebitda = 65
c_netdebt_ebitda = 70
c_de = 75
c_roe = 80
c_roa = 85
c_eps = 90
for i in np.arange(len(work_table.index)):
    for j in np.arange(5):
        work_table.iloc[i, c_pe + j] = work_table.iloc[i, c_mcap + j] / work_table.iloc[i, c_inc + j]
        work_table.iloc[i, c_ps + j] = work_table.iloc[i, c_mcap + j] / work_table.iloc[i, c_rev + j]
        work_table.iloc[i, c_pb + j] = work_table.iloc[i, c_mcap + j] / (work_table.iloc[i, c_tass + j] - work_table.iloc[i, c_tliab + j])
        work_table.iloc[i, c_ev_ebitda + j] = (work_table.iloc[i, c_mcap + j] + work_table.iloc[i, c_debt + j] - work_table.iloc[i, c_cash + j]) / work_table.iloc[i, c_ebitda + j]
        work_table.iloc[i, c_netdebt_ebitda + j] = (work_table.iloc[i, c_debt + j] - work_table.iloc[i, c_cash + j]) / work_table.iloc[i, c_ebitda + j]
        work_table.iloc[i, c_de +j] = work_table.iloc[i, c_debt + j] / (work_table.iloc[i, c_tass + j] - work_table.iloc[i, c_tliab + j])
        work_table.iloc[i, c_roe + j] = work_table.iloc[i, c_inc + j] / (work_table.iloc[i, c_tass + j] - work_table.iloc[i, c_tliab + j])
        work_table.iloc[i, c_roa + j] = work_table.iloc[i, c_inc + j] / work_table.iloc[i, c_tass + j]
        work_table.iloc[i, c_eps + j] = work_table.iloc[i, c_inc + j] * work_table.iloc[i, c_price + j] / work_table.iloc[i, c_mcap + j]
```
